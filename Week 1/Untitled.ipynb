```python
import numpy as np
from numpy.fft import fft
import pydub
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline
```

```python
song = Path('merah_putih.ogg')
a = pydub.AudioSegment.from_ogg(song)
y = np.array(a.get_array_of_samples())
```

```python
plt.plot(y)
plt.show()
```

```python
yf = fft(y)
plt.plot(np.abs(yf))
plt.show()
```
