```python
import numpy as np
from numpy.fft import fft
import pydub
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
import keras
from tensorflow.keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import SGD
%matplotlib inline
```

```python
song = Path('Rec/Kanan.wav')
a = pydub.AudioSegment.from_wav(song)
y = np.array(a.get_array_of_samples())
```

```python
plt.plot(y)
plt.show()
```

```python
yf = fft(y)
plt.plot(np.abs(yf))
plt.show()
```

```python
f = []
w = 256
for i in range(round(len(y)/w)-1):
    f.append(y[i*w:(i+1)*w-1])
len(f)
```

```python
# For positive range
data0 = []
data1 = []
data2 = []
data3 = []
data4 = []
data5 = []
data6 = []
data7 = []
data8 = []
```

```python
# For Negative range
data9 = []
data10 = []
data11 = []
data12 = []
data13 = []
data14 = []
data15 = []
data16 = []
data17 = []
```

```python
second_way = []
for i in range(len(f)):
    second_way.append(np.corrcoef(f[0],f[i])[0][1])
```

```python
second_way.sort()
second_way
```

```python
negative = []
positive = []
for i in second_way:
    if i < 0:
        negative.append(i)
    else:
        positive.append(i)
```

```python
for i in range(len(positive)):
    if i < 11:
        data0.append(positive[i])
    elif 11 <= i < 22:
        data1.append(positive[i])
    elif 22 <= i < 33:
        data2.append(positive[i])
    elif 33 <= i < 44:
        data3.append(positive[i])
    elif 44 <= i < 55:
        data4.append(positive[i])
    elif 55 <= i < 66:
        data5.append(positive[i])
    elif 66 <= i < 77:
        data6.append(positive[i])
    elif 77 <= i < 88:
        data7.append(positive[i])
    else:
        data8.append(positive[i])
```

```python
x2 = []
for i in range(len(f)):
    y = np.corrcoef(f[0],f[i])[0][1]
    if data0[0] <= y <= data0[10]:
        x2.append(0)
    elif data1[0] <= y <= data1[10]:
        x2.append(1)
    elif data2[0] <= y <= data2[10]:
        x2.append(2)
    elif data3[0] <= y <= data3[10]:
        x2.append(3)
    elif data4[0] <= y <= data4[10]:
        x2.append(4)
    elif data5[0] <= y <= data5[10]:
        x2.append(5)
    elif data6[0] <= y <= data6[10]:
        x2.append(6)
    elif data7[0] <= y <= data7[10]:
        x2.append(7)
    elif data8[0] <= y <= data8[24]:
        x2.append(8)
    elif data9[0] <= y <= data9[11]:
        x2.append(9)
    elif data10[0] <= y <= data10[11]:
        x2.append(10)
    elif data11[0] <= y <= data11[11]:
        x2.append(11)
    elif data12[0] <= y <= data12[11]:
        x2.append(12)
    elif data13[0] <= y <= data13[11]:
        x2.append(13)
    elif data14[0] <= y <= data14[11]:
        x2.append(14)
    elif data15[0] <= y <= data15[11]:
        x2.append(15)
    elif data16[0] <= y <= data16[11]:
        x2.append(16)
    elif data17[0] <= y <= data17[4]:
        x2.append(17)
```

```python
lb2 = to_categorical(x2)
```

```python
model = Sequential()
model.add(Dense(len(f[0]), activation='relu',input_shape=f[0].shape))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(f[0]), activation='relu'))
model.add(Dense(len(lb2[0,:]), activation='softmax'))
model.summary()
```

```python
lb2.shape
```

```python
model.compile(loss='categorical_crossentropy', optimizer="adam",metrics=['accuracy'])
```

```python
log = model.fit(np.array(f),np.array(lb2),validation_data=(np.array(f),np.array(lb2)),epochs=30,batch_size=32, verbose=True)
```

```python
plt.figure()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(log.history['loss'])
plt.plot(log.history['val_loss'])
plt.legend(['Training','Validation'])

plt.figure()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.plot(log.history['accuracy'])
plt.plot(log.history['val_accuracy'])
plt.legend(['Training','Validation'], loc='lower right')
plt.show()
```

```python
p = model.predict(np.array(f))
scores = model.evaluate(np.array(f),np.array(lb2))
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
for i in range(len(f)):
    print(f"{i}\n{p[i]}\n{lb2[i]}")
```
