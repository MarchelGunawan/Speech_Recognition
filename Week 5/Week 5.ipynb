```python
model = Sequential()
model.add(Dense(len(f[0]), activation='relu', input_shape=f[0].shape))
model.add(Dense(len(f[0]), activation='relu',))
model.add(Dense(len(f[0]), activation='relu',))
model.add(Dense(len(lb2[0,:]), activation='relu',))
model.summary()
```
