```python
from pydub import AudioSegment
import numpy as np
from numpy.fft import fft
import matplotlib.pyplot as plt
```

```python
a = AudioSegment.from_wav('Marchel Week10.wav')
y = np.array(a.get_array_of_samples())
```

```python
plt.plot(y)
plt.show()
```

```python
yf = fft(y)
plt.plot(np.abs(yf))
plt.show()
```
