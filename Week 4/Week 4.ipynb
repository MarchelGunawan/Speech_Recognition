```python
import numpy as np
from numpy.fft import fft
import pydub
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
import csv
%matplotlib inline
```

```python
song = Path('Michael_Ini Ibu Budi.wav')
a = pydub.AudioSegment.from_wav(song)
y = np.array(a.get_array_of_samples())
```

```python
plt.plot(y)
plt.show()
```

```python
yf = fft(y)
plt.plot(np.abs(yf))
plt.show()
```

```python
f = []
w = 256
for i in range(round(len(y)/w)-1):
    f.append(y[i*w:(i+1)*w-1])
len(f)
```

```python
np.savetxt("frm.csv",f,delimiter=",")
```

```python
second_way = []
for i in range(len(f)):
    second_way.append(np.corrcoef(f[0],f[i])[0][1])
```
